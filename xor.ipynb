```python
%matplotlib inline

import numpy as np
import tensorflow as tf
import tensorflow.keras as keras
import matplotlib.pyplot as plt
```

```python
input_xor = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
output_xor = np.array([[0], [1], [1], [0]])
input_xor.shape, output_xor.shape
```

```python
def create_model(hidden_size):
    model = keras.models.Sequential()
    model.add(keras.layers.Dense(hidden_size, activation='sigmoid', input_dim=input_xor.shape[1]))
    model.add(keras.layers.Dense(1, activation='sigmoid'))
    sgd = keras.optimizers.SGD(learning_rate=0.3)
    mse = keras.losses.MeanSquaredError()
    model.compile(loss=mse, optimizer=sgd)
    return model
```

```python
model_2 = create_model(2)
model_4 = create_model(4)
model_8 = create_model(8)
```

```python
nb_epoch = 10
history_2 = model_2.fit(input_xor, output_xor, epochs=nb_epoch, verbose=0)
history_4 = model_4.fit(input_xor, output_xor, epochs=nb_epoch, verbose=0)
history_8 = model_8.fit(input_xor, output_xor, epochs=nb_epoch, verbose=0)
```

```python
plt.figure(figsize=(25,5))
plt.subplot(131)
plt.plot(np.arange(nb_epoch), history_2.history['loss'])
plt.yscale('log')
plt.title("NN with hidden layer of 2. Loss for each epoch")
plt.subplot(132)
plt.plot(np.arange(nb_epoch), history_4.history['loss'])
plt.yscale('log')
plt.title("NN with hidden layer of 4. Loss for each epoch")
plt.subplot(133)
plt.plot(np.arange(nb_epoch), history_8.history['loss'])
plt.yscale('log')
plt.title("NN with hidden layer of 8. Loss for each epoch")
plt.show()
```

```python
print(f"Prediction from model 2:\n{model_2.predict(input_xor)}")
print(f"Prediction from model 4:\n{model_4.predict(input_xor)}")
print(f"Prediction from model 8:\n{model_8.predict(input_xor)}")
```
